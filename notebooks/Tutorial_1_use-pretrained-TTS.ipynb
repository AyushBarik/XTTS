````python
from scipy.io import wavfile
import numpy as np
from scipy.signal import butter, sosfilt
import pyloudnorm as pyln
import noisereduce as nr

# Load the WAV file
sample_rate, audio_data = wavfile.read("output.wav")
audio_data = audio_data.astype(np.float32) / np.max(np.abs(audio_data))  # Normalize to [-1, 1]

# Step 1: Noise Reduction
reduced_noise = nr.reduce_noise(y=audio_data, sr=sample_rate, prop_decrease=0.9)

# Step 2: High-pass filter to remove low-end rumble
sos_hp = butter(2, 100, btype='highpass', fs=sample_rate, output='sos')  # Cutoff at 100 Hz
filtered_audio = sosfilt(sos_hp, reduced_noise)

# Step 3: Gentle damping of high frequencies using FFT
freq = np.fft.rfftfreq(len(filtered_audio), 1 / sample_rate)
fft_data = np.fft.rfft(filtered_audio)

# Apply gentle damping to high frequencies above 5 kHz
high_damping_start = 5000  # Start reducing at 5 kHz
high_damping_end = 8000    # Strong reduction at 8 kHz
damping_curve = np.ones_like(freq)
damping_curve[freq > high_damping_start] = 1 - 0.4 * ((freq[freq > high_damping_start] - high_damping_start) /
                                                      (high_damping_end - high_damping_start))
damping_curve[freq > high_damping_end] = 0.6  # Damp everything above 8 kHz by 40%

# Apply the damping curve
fft_data *= damping_curve

# Transform back to time domain
eq_audio = np.fft.irfft(fft_data, n=len(filtered_audio))

# Step 4: Normalize loudness to -14 LUFS
meter = pyln.Meter(sample_rate)
loudness = meter.integrated_loudness(eq_audio)
normalized_audio = pyln.normalize.loudness(eq_audio, loudness, -14.0)

# Normalize peak to -3dB
normalized_audio = normalized_audio / np.max(np.abs(normalized_audio)) * 0.9

# Save the processed audio
from scipy.io.wavfile import write
write("output_refined.wav", sample_rate, (normalized_audio * 32767).astype(np.int16))

# Play the processed audio
import IPython
IPython.display.Audio("output_refined.wav")







````
